```python
from selenium.common.exceptions import NoSuchElementException

my_elems = mydriver.find_elements("xpath", '//div[@id="monthly_tab"]//div[@class="card  border-radius-RTB"]')

monthly_tab = []
failed = 0
i=0
while i < len(my_elems):
    try:
        record = {}

        record["plan_name"] = my_elem.find_element("xpath", './/h5[@class="plan-name blue-color mt-1"]').text
        if record["plan_name"] == "":
            mydriver.find_element("xpath", '//div[@id="monthly_tab"]//button[@class="slick-next slick-arrow"]').click()
            continue  # إذا الاسم فاضي، نتخطى السجل
        
        # نملأ باقي البيانات
        record["plan_price"] = my_elem.find_element("xpath", './/h5[@class="plan-price "]').text
        record["plan_hint"] = my_elem.find_element("xpath", './/span[@class="plan-hint mediumGrey-color"]').text

        # استخدمنا الـ <ul> اللي فيه الكلاس اللي قولتي عليه
        ul_elem = my_elem.find_element("xpath", './/ul[@class="list-group list-group-flush d-flex flex-column"]')
        list_items = ul_elem.find_elements("xpath", './li')

        # نملأ البيانات باستخدام فحص الطول علشان نتجنب IndexError
        record["included_mixes"] = list_items[0].text if len(list_items) > 0 else None
        record["favorite_numbers"] = list_items[1].text if len(list_items) > 1 else None
        record["renewal_reward"] = list_items[2].text if len(list_items) > 2 else None
        record["included_apps"] = list_items[3].text if len(list_items) > 3 else None
        record["family_offer"] = list_items[4].text if len(list_items) > 4 else None
        record["home_internet"] = list_items[5].text if len(list_items) > 5 else None
        record["included_subscriptions"] = list_items[6].text if len(list_items) > 6 else None
        record["activation_code"] = list_items[7].text if len(list_items) > 7 else None

        monthly_tab.append(record)

    except NoSuchElementException:
        print("Element not found, skipping this record.")
        failed += 1
    i += 1
# حفظ البيانات في ملف JSON
with open("etisalat_plans.json", "w", encoding="utf-8") as f:
    json.dump(monthly_tab, f, ensure_ascii=False, indent=4)

print(f"✅ Finished scraping. {failed} failed out of {len(my_elems)} total elements.")
```

```python
from selenium.common.exceptions import NoSuchElementException
import json
import time

weekly_tab = []
failed = 0
i = 0

my_elems = mydriver.find_elements("xpath", '//div[@id="weekly_tab"]//div[@class="card  border-radius-RTB"]')

while i < len(my_elems):
    my_elem = my_elems[i]
    try:
        record = {}

        # اسم الباقة
        plan_name = my_elem.find_element("xpath", './/h5[contains(@class, "plan-name")]').text
        record["plan_name"] = plan_name
        if plan_name == "":
            # دوسي على الزر Next لو الباقة مش واضحة
            mydriver.find_element("xpath", '//div[@id="weekly_tab"]//button[@class="slick-next slick-arrow"]').click()
            time.sleep(1)
            my_elems = mydriver.find_elements("xpath", '//div[@id="weekly_tab"]//div[@class="card  border-radius-RTB"]')
            continue

        # بيانات أخرى
        record["plan_hint"] = my_elem.find_element("xpath", './/h5[contains(@class,"plan-hint")]').text

        # الوصول إلى <ul> ثم <li> داخله
        ul_elem = my_elem.find_element("xpath", './/ul[contains(@class,"list-group-flush")]')
        list_items = ul_elem.find_elements("xpath", './li')

        # ملء البيانات مع التحقق
        record["included_mixes"] = list_items[0].text if len(list_items) > 0 else None
        record["included_apps"] = list_items[1].text if len(list_items) > 1 else None
        record["validity_period"] = list_items[2].text if len(list_items) > 2 else None
        record["recharge_code"] = list_items[3].text if len(list_items) > 3 else None
        record["renewal_code"] = list_items[4].text if len(list_items) > 4 else None
        record["social_megabytes"] = list_items[5].text if len(list_items) > 5 else None

        weekly_tab.append(record)

    except NoSuchElementException:
        print(f"Element not found in index {i}, skipping.")
        failed += 1

    i += 1

# حفظ البيانات في JSON
with open("etisalat_weekly_tab.json", "w", encoding="utf-8") as f:
    json.dump(weekly_tab, f, ensure_ascii=False, indent=4)

print(f"✅ Scraping weekly tab done. Extracted {len(weekly_tab)} cards. Failed: {failed}")
```

# connect tab

```python
from selenium.common.exceptions import NoSuchElementException
import time
import json

connect_tab = []
failed = 0
i = 0

my_elems = mydriver.find_elements("xpath", '//div[@class="tabsContainer-two"]//div[@class="card  border-radius-RTB"]')

while i < len(my_elems):
    my_elem = my_elems[i]
    try:
        record = {}

        # plan_name
        plan_name = my_elem.find_element("xpath", './/h5[contains(@class, "plan-name")]').text
        record["plan_name"] = plan_name
        if plan_name.strip() == "":
            # اضغط Next لو الاسم مش ظاهر
            mydriver.find_element("xpath", '//div[@class="tabsContainer-two"]//button[@class="slick-next slick-arrow"]').click()
            time.sleep(1)
            my_elems = mydriver.find_elements("xpath", '//div[@class="tabsContainer-two"]//div[@class="card  border-radius-RTB"]')
            continue

        # الوصول إلى <ul> ثم li العناصر
        ul_elem = my_elem.find_element("xpath", './/ul[contains(@class,"list-group-flush")]')
        list_items = ul_elem.find_elements("xpath", './li')

        # included_megabytes + plan_hint (داخل li الأول)
        if len(list_items) > 0:
            try:
                record["included_megabytes"] = list_items[0].find_element("xpath", './/p[contains(@class,"fs-16")]').text
            except:
                record["included_megabytes"] = None

            try:
                record["plan_hint"] = list_items[0].find_element("xpath", './/span[contains(@class,"plan-hint")]').text
            except:
                record["plan_hint"] = None
        else:
            record["included_megabytes"] = None
            record["plan_hint"] = None

        # etisalat_minutes
        record["etisalat_minutes"] = list_items[1].text if len(list_items) > 1 else None

        # plan_price من li الثالث
        if len(list_items) > 2:
            try:
                record["plan_price"] = list_items[2].find_element("xpath", './/h5[contains(@class,"plan-price")]').text
            except:
                record["plan_price"] = None
        else:
            record["plan_price"] = None

        connect_tab.append(record)

    except NoSuchElementException:
        print(f"Element not found in index {i}, skipping.")
        failed += 1

    i += 1

# حفظ في JSON
with open("etisalat_connect_tab.json", "w", encoding="utf-8") as f:
    json.dump(connect_tab, f, ensure_ascii=False, indent=4)

print(f"✅ Scraping connect tab done. Extracted {len(connect_tab)} plans. Failed: {failed}")
```

# extra_tab

```python
from selenium.common.exceptions import NoSuchElementException
import time
import json

extra_tab = []
failed = 0
i = 0

my_elems = mydriver.find_elements("xpath", '//div[@class="tabsContainersExtra"]//div[@class="card  border-radius-RTB"]')

while i < len(my_elems):
    my_elem = my_elems[i]
    try:
        record = {}

        # اسم الباقة
        plan_name = my_elem.find_element("xpath", './/h5[contains(@class, "plan-name")]').text
        record["plan_name"] = plan_name
        if plan_name.strip() == "":
            mydriver.find_element("xpath", '//div[@class="tabsContainersExtra"]//button[@class="slick-next slick-arrow"]').click()
            time.sleep(1)
            my_elems = mydriver.find_elements("xpath", '//div[@class="tabsContainersExtra"]//div[@class="card  border-radius-RTB"]')
            continue

        # السعر
        try:
            record["plan_price"] = my_elem.find_element("xpath", './/h5[contains(@class, "plan-price")]').text
        except:
            record["plan_price"] = None

        # hint (أسفل السعر)
        try:
            record["plan_hint"] = my_elem.find_element("xpath", './/span[contains(@class, "plan-hint")]').text
        except:
            record["plan_hint"] = None

        # الوصول إلى عناصر <ul><li>
        ul_elem = my_elem.find_element("xpath", './/ul[contains(@class,"list-group-flush")]')
        list_items = ul_elem.find_elements("xpath", './li')

        # extra_mixes
        if len(list_items) > 0:
            try:
                record["extra_mixes"] = list_items[0].find_element("xpath", './/p[contains(@class,"fs-16")]').text
            except:
                record["extra_mixes"] = None
        else:
            record["extra_mixes"] = None

        # activation_code
        if len(list_items) > 1:
            try:
                record["activation_code"] = list_items[1].find_element("xpath", './/p[contains(@class,"dir-ltr")]').text
            except:
                record["activation_code"] = None
        else:
            record["activation_code"] = None

        extra_tab.append(record)

    except NoSuchElementException:
        print(f"Element not found in index {i}, skipping.")
        failed += 1

    i += 1

# حفظ البيانات في JSON
with open("etisalat_extra_tab.json", "w", encoding="utf-8") as f:
    json.dump(extra_tab, f, ensure_ascii=False, indent=4)

print(f"✅ Scraping extra tab done. Extracted {len(extra_tab)} plans. Failed: {failed}")
```

# exchange_tab

```python
from selenium.common.exceptions import NoSuchElementException
import time
import json

exchange_tab = []
failed = 0
i = 0

my_elems = mydriver.find_elements("xpath", '//section[@id="for_textContainer"]//div[contains(@class,"text-container")]')

while i < len(my_elems):
    my_elem = my_elems[i]
    try:
        record = {}

        # total_gb
        try:
            total_text = my_elem.find_element("xpath", './/div[@class="title"]/h5').text
            record["total_gb"] = total_text
        except:
            record["total_gb"] = None

        # description
        try:
            record["description"] = my_elem.find_element("xpath", './/div[@class="title"]/p').text
        except:
            record["description"] = None

        # available_for
        try:
            record["available_for"] = my_elem.find_element("xpath", './/div[contains(@class, "my-3")]/h6').text
        except:
            record["available_for"] = None

        # daytime_gb and note
        try:
            daytime_section = my_elem.find_elements("xpath", './/div[@class="my-3 d-flex justify-content-between"]')[0]
            record["daytime_gb"] = daytime_section.find_element("xpath", './/h6').text
            record["daytime_note"] = daytime_section.find_element("xpath", './/p').text
        except:
            record["daytime_gb"] = None
            record["daytime_note"] = None

        # nighttime_gb and note
        try:
            nighttime_section = my_elem.find_elements("xpath", './/div[@class="my-3 d-flex justify-content-between"]')[1]
            record["nighttime_gb"] = nighttime_section.find_element("xpath", './/h6').text
            record["nighttime_note"] = nighttime_section.find_element("xpath", './/p').text
        except:
            record["nighttime_gb"] = None
            record["nighttime_note"] = None

        exchange_tab.append(record)

    except NoSuchElementException:
        print(f"Element not found in index {i}, skipping.")
        failed += 1

    i += 1

# حفظ النتائج
with open("etisalat_exchange_tab.json", "w", encoding="utf-8") as f:
    json.dump(exchange_tab, f, ensure_ascii=False, indent=4)

print(f"✅ Done scraping exchange section: {len(exchange_tab)} items scraped, {failed} failed.")
```
